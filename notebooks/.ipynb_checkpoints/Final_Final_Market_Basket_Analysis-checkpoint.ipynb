```python
item_list = items['StockCode'].tolist()
```

```python
from mlxtend.preprocessing import TransactionEncoder

item_list = [[str(item) for item in transaction] for transaction in item_list]

te = TransactionEncoder()
te_ary = te.fit(item_list).transform(item_list)
encoded_df = pd.DataFrame(te_ary, columns=te.columns_)
encoded_df.head()
```

# Clustering

```python
import matplotlib.pyplot as plt

inertia = []
for i in range(1, 11):  # Checking for k from 1 to 10
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(encoded_df)
    inertia.append(kmeans.inertia_)

plt.plot(range(1, 11), inertia)
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method')
plt.show()
```

### Confidence and lift Eval

##### 1) Confidence 
    -->is nothing but the confidence of having B after having A
confidence(Aâ‡’B)= support(AâˆªB) / support(A)

##### 2) Lift
    -->is nothing but the Lift provided by A to the confidence of having B
â€‹lift(Aâ‡’B) = support(AâˆªB) / [support(A)*support(B)]

```python
# Function to calculate confidence
def calc_confidence(antecedent, consequent, itemsets_df):
    support_antecedent = itemsets_df[itemsets_df['itemsets'] == antecedent]['support'].values[0]
    support_antecedent_consequent = itemsets_df[itemsets_df['itemsets'] == (antecedent | consequent)]['support'].values[0]
    confidence = support_antecedent_consequent / support_antecedent
    return confidence

# Function to calculate lift
def calc_lift(antecedent, consequent, itemsets_df):
    support_antecedent = itemsets_df[itemsets_df['itemsets'] == antecedent]['support'].values[0]
    support_consequent = itemsets_df[itemsets_df['itemsets'] == consequent]['support'].values[0]
    support_antecedent_consequent = itemsets_df[itemsets_df['itemsets'] == (antecedent | consequent)]['support'].values[0]
    lift = support_antecedent_consequent / (support_antecedent * support_consequent)
    return lift
```

```python
# Initialize lists to store results
confidence_list = []
lift_list = []
antecedent_list = []
consequent_list = []

# Loop over each 2-item itemset
for _, row in length_2_itemsets.iterrows():
    itemset = row['itemsets']
    items = list(itemset)  # Convert frozenset to list
    
    # Split into antecedent and consequent
    antecedent = frozenset([items[0]])
    consequent = frozenset([items[1]])
    
    # Calculate confidence and lift
    confidence = calc_confidence(antecedent, consequent, freq_itemsets_fpg)
    lift = calc_lift(antecedent, consequent, freq_itemsets_fpg)
    
    # Append to lists
    confidence_list.append(confidence)
    lift_list.append(lift)
    antecedent_list.append

length_2_itemsets['confidence'] = confidence_list
length_2_itemsets['lift'] = lift_list
length_2_itemsets.describe()
```

# Final Dataset

```python
high_confidence.head()
```

```python
item_names = df.drop(columns='InvoiceNo')
item_names['StockCode'] = item_names['StockCode'].astype(str)
item_names = item_names.drop_duplicates(subset=['StockCode'])
item_names.head()
```

```python
# Step 1: Create a new column 'itemsets' that contains tuples of the original itemset pairs
high_confidence['itemsets_tuple'] = high_confidence['itemsets'].apply(lambda x: tuple(x) if isinstance(x, list) else x)

# Step 2: Explode the itemsets so we work with individual items
high_confidence_exploded = high_confidence.explode('itemsets')

# Step 3: Merge item descriptions with exploded itemsets
high_confidence_exploded = pd.merge(high_confidence_exploded, 
                                     item_names[['StockCode', 'Description']],
                                     left_on='itemsets', 
                                     right_on='StockCode', 
                                     how='left')

# Step 4: Group by 'support', 'confidence', 'lift', and the 'itemsets_tuple' to keep itemset pairs
high_confidence_exploded_grouped = high_confidence_exploded.groupby(['support', 'confidence', 'lift', 'itemsets_tuple']).agg(
    itemsets_with_descriptions=('Description', lambda x: ', '.join(x))
).reset_index()

# Step 5: Rename columns to final output format
high_confidence_exploded_grouped.rename(columns={'itemsets_tuple': 'itemsets', 'itemsets_with_descriptions': 'itemsets_with_descriptions'}, inplace=True)

# Final output with itemsets and corresponding descriptions
high_confidence_final = high_confidence_exploded_grouped

high_confidence_final.head()
```

```python
high_confidence_final.loc[0, 'itemsets_with_descriptions']
```

# Visualization

```python
# Network Graphs

import networkx as nx
import matplotlib.pyplot as plt

# Create graph
G = nx.Graph()

for _, row in high_confidence_final.iterrows():
    itemsets = row['itemsets_with_descriptions'].split(', ')
    for i in range(len(itemsets)):
        for j in range(i+1, len(itemsets)):
            if not G.has_edge(itemsets[i], itemsets[j]):
                G.add_edge(itemsets[i], itemsets[j], weight=row['lift'])

# Draw the graph with edge weight influencing width
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_size=500, node_color="lightblue", font_size=10, font_weight='bold')
edge_weights = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_weights)
plt.show()
```
